**Bar Plot by using Plotly**

**PX:plotly express is the easy-to-use, high level interface to plotly,which operates on a variety of types of data and produces easy-to-style figures.**

```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv("/content/Cardekho.csv")
```

```python
df.head()
```

```python
df.tail()
```

```python
df["max_price"].value_counts()
```

```python
px.bar(df, x="max_price", color="max_price")
```

```python
px.bar(df, x="max_price", color="max_price",
       color_discrete_sequence=["red", "blue", "green", "purple", "orange"])
```

```python
data = df.groupby(["max_price"]).size().rename("count").reset_index()
data.head()
```

```python
px.bar(data, x="max_price", y="count", color="max_price", text="count")
```

**Box plot by using plotly**

```python
import plotly.express as px
```

```python
df = px.data.tips()
df.head()
```

```python
fig = px.box(df, y= "total_bill")
fig.show()
```

```python
fig = px.box(df, x= "time", y= "total_bill", points= "outliers")
fig.show()
```

```python
fig = px.box(df, x= "time", y="total_bill", points= "all")
fig.show()
```

```python
fig = px.box(df, x= "time", y="total_bill", points= "outliers",
             color= "smoker")
fig.show()
```

```python
fig = px.box(df, x= "time", y="total_bill", points= "outliers",
             color= "smoker", notched= True,
             title= "box plot of total bill",
             hover_data = ["day"]
             )
fig.show()
```

```python
fig = px.strip(df, x= "day", y="tip")
fig.show()
```

**Violin plot in plotly**

```python
import plotly.express as px
```

```python
df = px.data.tips()
df.head()
```

```python
fig = px.violin(df, y= "total_bill")
fig.show()
```

```python
fig = px.violin(df, y="total_bill", box= True,
                points= "all")
fig.show()
```

```python
fig = px.violin(df, y="total_bill", box= False,
                points= "all")
fig.show()
```

```python
df.columns
```

```python
fig = px.violin(df, y= "tip", color= "time", violinmode= "overlay",
                hover_data= df.columns)
fig.show()
```

```python
fig = px.violin(df, y= "total_bill")
fig.update_traces(quartilemethod= "exclusive")
fig.show()
```

```python
fig = px.violin(df, y= "total_bill")
fig.update_traces(quartilemethod= "inclusive")
fig.show()
```
